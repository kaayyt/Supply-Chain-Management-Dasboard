```python
import pandas as pd

# Load the uploaded file into a pandas DataFrame
data = pd.read_csv('supply_chain_data.csv')
# Display the first few rows
data.head()
```

```python
# Display basic information about the data
print(data.info())

# Show descriptive statistics for numeric columns
data.describe()
```

```python
# Check for missing values
missing_data = data.isnull().sum()
print("Missing values:\n", missing_data)
```

```python
# Define the column name to check for outliers
column_name = 'Price'  # Replace with your actual column name

# Check if the column exists before processing
if column_name in data.columns:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = data[column_name].quantile(0.25)
    Q3 = data[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Filter out rows that are outliers
    data = data[(data[column_name] >= Q1 - 1.5 * IQR) & (data[column_name] <= Q3 + 1.5 * IQR)]
else:
    print(f"Column '{column_name}' not found in the dataset.")
```

```python
# Save cleaned data
cleaned_file_path = 'cleaned_supply_chain_data.csv'
data.to_csv(cleaned_file_path, index=False)

# Download the cleaned file
files.download(cleaned_file_path)
```
